```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
ur1 = "http://www.tianqihoubao.com/weather/top/guangzhou.html"
response = requests.get(ur1)
```

```python
soup = BeautifulSoup(response.text, 'html.parser') 
```

```python
soup
```

```python
wdetail = soup.find("div", {"class": "wdetail"})
tr_all=wdetail.find_all("tr")
data_all = []
for i in tr_all:
    td_all=i.find_all("td")
    data = []
    for j in td_all:
        data.append(j.text)
    data_all.append(data)
       
```

```python
data_all
```

```python
weather=pd.DataFrame(data_all)
weather.columns=["城市","日期","天气状况","风力方向","最高温度","天气状况","风力方向","最低温度"]
weather.drop([0],inplace=True)
weather.to_csv("xxx.csv",encoding="utf_8_sig")
```
